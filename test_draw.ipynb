```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/economic_data.csv")

important_df = df[df['importance'] == 'high']
important_df['date']= pd.to_datetime(important_df['date'], format='%d/%m/%Y')

cpi_data = important_df[important_df['event'].str.contains(r'^GDP (QoQ)')]
display(cpi_data.head())
# グラフの描画
plt.figure(figsize=(20, 5))
plt.plot(cpi_data['date'], cpi_data['forecast'], label='Forecast')
plt.plot(cpi_data['date'], cpi_data['actual'], label='Actual')
plt.title('CPI Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('CPI Change')
plt.legend()
plt.grid(True)
plt.show()
```
